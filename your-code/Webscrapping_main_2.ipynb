#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url13 = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
#your code
response13 = requests.get(url13)
response13
````

````python
soup13 = BeautifulSoup(response13.content)
````

````python
bloque13 = soup13.select('table > tbody')
#bloque13[0].text.strip(' \ n')
bloque_todo = bloque13[0].select('td>a')
bloque_todo
````

````python
bloque_branch = bloque13[0].select('td:last-child')
bloque_branch[0].text.strip('\n')
lista_branch = [bloque.text.strip('\n') for bloque in bloque_branch]
lista_branch
````

````python
bloque_language = bloque13[0].select('td:first-child')
bloque_language[0].text.strip('\n')
lista_language = [bloque.text.strip('\n') for bloque in bloque_language]
lista_language
````

````python
bloque_speakers = bloque13[0].select('td:nth-child(2)')
bloque_speakers[0].text.strip('\n')
lista_speakers = [bloque.text.strip('\n') for bloque in bloque_speakers]
lista_speakers
````

````python
bloque_family = bloque13[0].select('td:nth-child(3)')
bloque_family[0].text.strip('\n')
lista_family = [bloque.text.strip('\n') for bloque in bloque_family]
lista_family
````

````python
df_languages = pd.DataFrame(lista_language)
df_languages.rename(columns= {0:'Language'}, inplace = True)
df_languages['Native speakers (millions)'] = lista_speakers
df_languages['Language family'] = lista_family
df_languages['Branch'] = lista_branch
df_languages[:10]
````
